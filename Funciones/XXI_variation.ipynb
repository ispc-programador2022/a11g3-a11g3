```python
def varianza(lista_numeros):
    med = media(lista_numeros) #llamo a la funcion media
    n = len(lista_numeros)
    suma = 0
    for dato in lista_numeros:
        suma += (dato-med)**2
    varianza = suma/n
    return varianza
```
